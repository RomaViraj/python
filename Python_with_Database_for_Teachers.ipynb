Firebase is developed by Google in 2012. It is is a database to store and synchronize data in real-time. **SQLite3** is a very easy to use database engine. It is self-contained, serverless, zero-configuration and transactional. It is very fast and lightweight, and the entire database is stored in a single disk file. It is used in a lot of applications as internal data storage. The Python Standard Library includes a module called "sqlite3" intended for working with this database. 

**SQLITE vs MYSQL**

This is also referred to as an embedded database which means the DB engine runs as a part of the app. On the other hand, MySQL requires a server to run. MySQL will require a client and server architecture to interact over a network.

**The key differences between Firebase and MySQL:**

**Architecture:** Firebase is a NoSQL database that stores and syncs data in real-time (a real-time document store)
